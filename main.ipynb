```python
#Hacer un programa que pida al usuario dos números y muestre por pantalla su división. Si el divisor es cero, el programa debe mostrar un error

n = float(input("Introducir el dividendo: "))
m = float(input("Introducir el divisior: "))
if m == 0:
    print("¡Error! No se puede dividir por el 0")
else:
    print(n/m)
```

```python
#Para pagar impuestos una persona debe ser mayor de 18 años y tener ingresos superiores a $12,000
#Hacer un programa que pregunte al usuario su edad y sus ingresos mensuales y muestre por pantalla si tiene que pagar impuestos o no

age = int(input("¿Cuántos años tienes? "))
if age < 18: 
    print ("No puedes igresar, eres menor de edad")
else:
    print("Eres mayor de edad, pasa a la siguiente pregunta")
ingresos = float(input("¿Cuáles son tus ingresos mensuales? "))
if ingresos >= 12000:
    print("Tienes que pagar impuestos")
else:
    print("No tienes que pagar impuestos")
```
